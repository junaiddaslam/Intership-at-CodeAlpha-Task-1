```python
import random

def hangman():
    # List of words for the game
    words = ["python", "hangman", "challenge", "programming", "internship"]
    
    # Select a random word from the list
    word = random.choice(words)
    
    # Initialize the guessed word with underscores
    guessed = "_" * len(word)
    
    # Convert the word and guessed word to lists for easier manipulation
    word = list(word)
    guessed = list(guessed)
    
    # List to keep track of letters that have been guessed
    letters_guessed = []
    
    # Number of attempts the player has
    attempts = 6

    # Game loop
    while attempts > 0:
        # Print the current state of the guessed word
        print(' '.join(guessed))
        
        # Get a letter guess from the player
        guess = input("Guess a letter: ").lower()

        # Check if the letter has already been guessed
        if guess in letters_guessed:
            print("You already guessed that letter.")
        
        # Check if the guessed letter is in the word
        elif guess in word:
            # Update the guessed word with the correct letter
            for index, letter in enumerate(word):
                if letter == guess:
                    guessed[index] = guess
            letters_guessed.append(guess)
        else:
            # Decrease the number of attempts if the guess is incorrect
            attempts -= 1
            print(f"Wrong guess. You have {attempts} attempts left.")
            letters_guessed.append(guess)

        # Check if the word has been completely guessed
        if "_" not in guessed:
            print("Congratulations! You guessed the word:", ''.join(guessed))
            break
    else:
        # Print the word if the player runs out of attempts
        print("Sorry, you ran out of attempts. The word was:", ''.join(word))

if __name__ == "__main__":
    hangman()
```

```python
import requests

class StockPortfolio:
    def __init__(self):
        self.stocks = {}  # Dictionary to hold stock symbols and the number of shares

    def add_stock(self, symbol, shares):
        # Add or update the number of shares for a given stock symbol
        self.stocks[symbol] = self.stocks.get(symbol, 0) + shares

    def remove_stock(self, symbol, shares):
        # Remove shares of a given stock symbol
        if symbol in self.stocks:
            self.stocks[symbol] -= shares
            if self.stocks[symbol] <= 0:
                del self.stocks[symbol]

    def get_stock_price(self, symbol):
        api_key = '9X6nU0D9rdLMaTAYPEeZX05M9p8FVdgt'  # Your actual API key
        api_url = f'https://financialmodelingprep.com/api/v3/quote/{symbol}?apikey={api_key}'
        try:
            response = requests.get(api_url)
            response.raise_for_status()  # Check if the request was successful
            data = response.json()
            # Check if data is a non-empty list and contains 'price'
            if data and isinstance(data, list) and len(data) > 0 and 'price' in data[0]:
                return data[0]['price']
            else:
                print(f"Could not fetch price for {symbol}")
                return None
        except requests.exceptions.RequestException as e:
            print(f"An error occurred: {e}")
            return None

    def get_portfolio_value(self):
        total_value = 0  # Initialize total portfolio value
        # Calculate total value based on current stock prices
        for symbol, shares in self.stocks.items():
            price = self.get_stock_price(symbol)
            if price is not None:
                total_value += shares * price
        return total_value

if __name__ == "__main__":
    portfolio = StockPortfolio()
    portfolio.add_stock('AAPL', 10)
    portfolio.add_stock('GOOGL', 5)
    print("Portfolio Value: $", portfolio.get_portfolio_value())
```
